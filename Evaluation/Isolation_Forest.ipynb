# Training of model

```python
import pandas as pd
import re
from sklearn.ensemble import IsolationForest
import numpy as np
import pickle
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score,  recall_score, classification_report


# Define the path to your CSV file
file_path = 'train_normal.csv'

# Read the CSV into a DataFrame
df = pd.read_csv(file_path)
```

```python
from utils import preprocess_dataframe

df = preprocess_dataframe(df)
df.head()
```

```python
train = pd.get_dummies(df)

train.head()
```

```python
isolation_forest = IsolationForest(contamination='auto', random_state=42)
isolation_forest.fit(train)
```

# Evaluation

```python
# Define the path to your CSV file
file_path = 'test_combined.csv'

# Read the CSV into a DataFrame
test = pd.read_csv(file_path)
```

```python
test.head()
```

```python
test_x = test.drop(['gold truth'], axis=1)
test_y = test['gold truth']
```

```python
from utils import create_dummies
test_x = create_dummies(test_x)
test_x.head()
```

```python
scores = pd.DataFrame(isolation_forest.decision_function(test_x), columns=['score'])
```

```python
scores
```

```python
from sklearn.metrics import fbeta_score

def find_best_threshold(scores, true_labels, beta=2.0):
    best_threshold = None
    best_fbeta = 0

    # Define the range of thresholds to explore
    thresholds = np.linspace(scores['score'].min(), scores['score'].max(), 1000)
    
    for threshold in thresholds:
        predictions = scores['score'] < threshold
        current_fbeta = fbeta_score(true_labels, predictions, beta=beta)
        
        if current_fbeta > best_fbeta:
            best_fbeta = current_fbeta
            best_threshold = threshold

    return best_threshold, best_fbeta

# Use a beta value larger than 1 (e.g., beta=2) to prioritize recall
best_threshold, best_fbeta = find_best_threshold(scores, test_y, beta=2.0)
```

```python
# Assuming you already have best_predictions calculated
best_predictions = scores['score'] < best_threshold

# Calculate accuracy
accuracy = accuracy_score(test_y, best_predictions)

# Calculate AUC
auc = roc_auc_score(test_y, -scores['score'])

# Calculate recall
recall = recall_score(test_y, best_predictions)

print(f'Best Threshold: {best_threshold}')
print(f'Best F-beta Score: {best_fbeta:.4f}')
print(f'Accuracy: {accuracy:.4f}')
print(f'AUC Score: {auc:.4f}')
print(f'Recall: {recall:.4f}')
print(classification_report(test_y, best_predictions, target_names=['Normal', 'Attack']))
```
